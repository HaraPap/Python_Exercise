# **Data Frame**

A DataFrame represents a rectangular table of data and contains an ordered collection of columns. You can think of it as a spreadsheet or SQL table where each column has a column name for reference and each row can be accessed by using row numbers. Column names and row numbers are known as column and row index.

DataFrame is a fundamental Pandas data structure in which each column can be of a different value type (numeric, string, boolean, etc.). A data set can be first read into a DataFrame and then various operations (i.e. indexing, grouping, aggregation etc.) can be easily applied to it.

## **General Statistics Opertaions**

Pandas provide basic and advanced statitsics.

The agg() method allows you to apply a function or a list of function names to be executed along one of the axis of the DataFrame, default 0, which is the index (row) axis.

We will find mean, median, standard deviation, quantile , minimum and maximum for each arithmetic column.

```python
df.LATITUDE.agg(['mean' , 'median' , 'std' , 'quantile' , 'min' , 'max' ])
```

```python
df.LONGΤITUDE.agg(['mean' , 'median' , 'std' , 'quantile' , 'min' , 'max' ])
```

```python
df.agg(['min' , 'max' ]) # statistics to the dataframe as a whole
```

## **Maps**

GeoPandas is an open source project to make working with geospatial data in python easier. GeoPandas extends the datatypes used by pandas to allow spatial operations on geometric types. Geometric operations are performed by shapely. Geopandas further depends on pyogrio for file access and matplotlib for plotting.

```python
import geopandas as gpd
```

```python
from shapely.geometry import Point
```

```python
import plotly.express as px
```

**Create a new column with the coordinates**

```python
df_geo = gpd.GeoDataFrame(df, geometry= gpd.points_from_xy(df.LONGΤITUDE, df.LATITUDE), crs="EPSG:4326")  # WGS84 CRS
```

```python
df_geo
```

Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library.

```python
import folium 
```

```python
df_geo.explore() # explore() function gives us a map the we can scroll over and see all the metadata for each fort
```

```python
import warnings
warnings.filterwarnings('ignore')
```

We can seperate with comma the values in geometry if we want.

```python
df_geo["formatted_geometry"] = df_geo.geometry.apply(lambda point: f"POINT ({point.x},{point.y})")
```
